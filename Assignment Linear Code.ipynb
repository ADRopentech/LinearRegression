```python
# Declare Library:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
from sklearn import linear_model as LR
```

```python
# read file and store in Variable Data
file_path = r'C:\Users\s423160\OneDrive - Cranfield University\Desktop\Python\Assignment\diabetes.csv'

data = pd.read_csv(file_path, header = None)

headers = ['x', 'y']
data.columns = headers
```

```python
#Take 2 variable and store data as list in those 2 variables
x_columnlist , y_columnlist = list(data['x']), list(data['y'])
```

```python
#calculate Mean
x_bar = np.mean(x_columnlist)
y_bar = np.mean(y_columnlist)
```

```python
#calculate slope and intercept
slope = sum((x-x_bar)*(y-y_bar) for x,y in zip(x_columnlist,y_columnlist))/sum((x-x_bar)**2 for x in x_columnlist)
intercept = y_bar - (slope*x_bar)
```

```python
#create linear regression visualization 
plot.style.use('_mpl-gallery')

#setting up the Visualization
fig, ax = plot.subplots() 
fig.set_size_inches(8,6)
ax.set_xlim(-0.150, 0.200)
ax.set_ylim(0, 400)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Linear regression')

#Visualizing the datapoints in diabetes.csv
ax.scatter(x_columnlist, y_columnlist, label = 'data point', color = 'green')

#Visualizing the Linear Regression Line for the data

##Creating plotting dataset
x_values = np.linspace(-0.100, 0.150, 250)
y_values = slope * x_values + intercept

##Plotting Linear regression line
ax.plot(x_values, y_values, label='calculated regression line', color='red' )

#Validating Result against predefined Linear Regression Library in SciKit-Learn
X = np.array(x_columnlist).reshape(-1,1)  ##Input pameter in Linear Regression function
model = LR.LinearRegression()             ##Creating LR Model
model.fit(X, y_columnlist)                ##fitting Model with data

##Plotting Linear regression Plot from SciKit-Learn
ax.plot(x_columnlist, model.predict(X), label='Validation line', color = 'blue')

plot.show()
```
